```python
# Imports
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.preprocessing import StandardScaler
```

```python
# Reading the file
data = pd.read_csv('/content/drive/MyDrive/bank-data/bank-full.csv', delimiter=';')
data
```

```python
data.shape
```

```python
# replacing categorical features with numerical values
data['job'] = data['job'].map({'admin.':0, 'blue-collar':1, 'entrepreneur':2, 'housemaid':3, 'management':4,
                               'retired':5, 'self-employed':6, 'services':7, 'student':8, 'technician':9,
                               'unemployed':10, 'unknown':11})
data['marital'] = data['marital'].map({'divorced':0, 'married':1, 'single':2, 'unknown':3})
data['education'] = data['education'].map({'primary':0, 'secondary':1, 'tertiary':2, 'unknown':3})
data['default'] = data['default'].map({'no':0, 'yes':1, 'unknown':2})
data['housing'] = data['housing'].map({'no':0, 'yes':1, 'unknown':2})
data['loan'] = data['loan'].map({'no':0, 'yes':1, 'unknown':2})
data['contact'] = data['contact'].map({'cellular':0, 'telephone':1, 'unknown':2})
data['month'] = data['month'].map({'jan':0, 'feb':1, 'mar':2, 'apr':3, 'may':4, 'jun':5, 
                                   'jul':6, 'aug':7, 'sep':8, 'oct':9, 'nov':10, 'dec':11})
data['poutcome'] = data['poutcome'].map({'failure':0, 'success':1, 'other':2, 'unknown':3})
data['y'] = data['y'].map({'no':0, 'yes':1})

data.head()
```

```python
# separating features and target
X = data.drop('y', axis=1)
y = data['y']
```

```python
# splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# scaling the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
data_feat = pd.DataFrame(X_test,columns=data.columns[:-1])
data_feat.head()
```

```python
# Creating and training the classifier using 'sigmoid' kernel SVM
classifier = SVC(kernel='sigmoid', random_state=42)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
# calculate accuracy score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
print()
```

```python
cr = classification_report(y_test, y_pred)
print("Classification Report:")
print(cr)
```
